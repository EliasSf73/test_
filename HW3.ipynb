**(c) Perform best subset selection on the training set, and plot the training set MSE associated with the best model of each size (You may use the code from Lab 5) (Run as many number of features as your PC allows.).**

`````python
# !pip install tqdm

import statsmodels.api as sm
from tqdm import tqdm
from operator import itemgetter
from itertools import combinations

def best_subsets(dataframe, dataframe_test, predictors, response, max_features=8):

    def process_linear_model(features):

        # Create design Matrix
        X = sm.add_constant(dataframe[features])
        y = dataframe[response]

        model = sm.OLS(y,X).fit()
        RSS = model.ssr
        return (model, RSS, features)

    def get_best_kth_model(k):

        results = []

        for combo in combinations(predictors, k):
            # process linear model with this combo of features
            results.append(process_linear_model(list(combo)))

        # sort the models and return the one with the smallest RSS
        return sorted(results, key=itemgetter(1)).pop(0)

    models =[]
    train_mse = []
    test_mse = []
    for k in tqdm(range(1,max_features+1)):
        best_model, best_rss, best_combo = get_best_kth_model(k)
        n = len(best_model.model.endog)
        train_mse.append(best_rss / n)

        x_test = sm.add_constant(dataframe_test[best_combo])
        y_test = dataframe_test[response]
        predictions = best_model.predict(x_test)
        mse = ((y_test - predictions) ** 2).mean()
        test_mse.append(mse)
        models.append(best_model)

    return models, train_mse, test_mse
feature_names = [f'feature_{i}' for i in range(1, p+1)]
# print(feature_names)
df_train = pd.DataFrame(np.concatenate([x_train, y_train], axis=1), columns=feature_names + ['label'])
df_test = pd.DataFrame(np.concatenate([x_test, y_test], axis=1), columns=feature_names + ['label'])
models, train_mse, test_mse = best_subsets(df_train, df_test, feature_names, 'label', max_features=6)

# plot task
n_features = list(range(1, len(train_mse) + 1))

# Plotting the Test Set MSE
plt.figure(figsize=(10, 6))
plt.plot(n_features, train_mse, marker='o', linestyle='-', color='r')
plt.title('Train Set MSE by Model Size')
plt.xlabel('Number of Features')
plt.ylabel('Train Set MSE')
plt.grid(True)
plt.show()
`````

# 10. Backward stepwise selection (30pt)

**In our Lecture 5 and Lab 5, we learned about backward stepwise selection. Implement backward stepwise selection by referring to the implementation of forward stepwise selection from Lab 5. As a result of the function, 1) print the best model (predictors) for each $k$, 2) perform cross validation, and 3) plot CV errors for each $k$ as shown in Section 1.3 of Lab 5. Use ```Hitters``` dataset for this problem.**

`````python

def backward_stepwise_selection(data, response, predictors, n_splits=10):
    remaining_predictors = predictors[:]
    models = []
    results = {
        'features': [],
        'CV_errors': []
    }

    kf = KFold(n_splits=n_splits, shuffle=True, random_state=1)

    while len(remaining_predictors) > 0:
        best_aic = float('inf')
        best_model = None
        best_predictors = None

        for combo in combinations(remaining_predictors, len(remaining_predictors)-1):
            X = sm.add_constant(data[list(combo)])
            model = sm.OLS(data[response], X).fit()

            if model.aic < best_aic:
                best_aic = model.aic
                best_model = model
                best_predictors = combo

        remaining_predictors = list(best_predictors)

        models.append(best_model)
        results['features'].append(best_predictors)

        CV_errors = []
        for train_index, test_index in kf.split(data):
            cv_train, cv_test = data.iloc[train_index], data.iloc[test_index]
            model = sm.OLS(cv_train[response], sm.add_constant(cv_train[list(best_predictors)])).fit()
            predictions = model.predict(sm.add_constant(cv_test[list(best_predictors)]))
            mse = ((cv_test[response] - predictions) ** 2).mean()
            CV_errors.append(mse)

        results['CV_errors'].append(np.mean(CV_errors))

    # Plotting CV errors
    plt.figure(figsize=(10, 6))
    plt.plot(range(len(results['CV_errors'])), results['CV_errors'], marker='o', linestyle='-')
    plt.xticks(range(len(results['CV_errors'])))
    plt.xlabel('Number of Predictors')
    plt.ylabel('Cross-Validated MSE')
    plt.title('Backward Stepwise Selection CV Error')
    plt.gca().invert_xaxis()
    plt.show()

    return results, models

data = pd.read_csv('Hitters.csv').dropna()
dummies = pd.get_dummies(data[['League', 'Division', 'NewLeague']])
dummies = dummies.astype(int)
df = data.drop(['League', 'Division', 'NewLeague'], axis=1)
df = pd.concat([df, dummies[['League_N', 'Division_W', 'NewLeague_N']]],axis=1)

response = 'Salary'
predictors = [col for col in df.columns if col != response]

results, models = backward_stepwise_selection(df, response, predictors)
for idx, features in enumerate(results['features']):
    print(f"Best model with {len(features)} predictors: {features}")
    print(f"Cross-validated MSE: {results['CV_errors'][idx]}")
`````
