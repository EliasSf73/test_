Mean squared Error

```python
def loss_function(m,b,points):
  error_sum=0
  for i in range(len(points)):
    x=points.iloc[i].YearsExperience
    y=points.iloc[i].Salary
    error_sum+=(y-(m*x+b))**2
  error_sum/float(len(points))
```

Gradient Descent

```python
def gradient_descent(m_cur,b_cur,points,L):
  m_gradient=0
  b_gradient=0
  n=len(points)
  for i in range(n):
    x=points.iloc[i].YearsExperience
    y=points.iloc[i].Salary
    m_gradient+=-(2/n)*x*(y-(m_cur*x+b_cur))
    b_gradient+=-(2/n)*(y-(m_cur*x+b_cur))

  m=m_cur-m_gradient*L
  b=b_cur-b_gradient*L
  return m,b

m=0
b=0
L=0.0001
epochs=300
for i in range(epochs):
  if i%50==0:
    print(f"epoch:{i}")
  m,b=gradient_descent(m,b,data,L)
print(m,b)

plt.scatter(data.YearsExperience,data.Salary,color='red')
plt.plot(list(range(6,17)),[m*x+b for x in range(6,17)],color='blue')
plt.show()
```
